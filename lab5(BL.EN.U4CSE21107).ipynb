```python
#A7

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [1]])

def sigmoid_activation(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    return sigmoid_activation(z) * (1 - sigmoid_activation(z))

def neural_network(X, y, alpha):
    # Define the initial weights and biases
    w1 = np.array([[0.1, 0.2], [0.3, 0.4]])
    b1 = np.array([[0.5], [0.6]])
    w2 = np.array([[0.7], [0.8]])
    b2 = np.array([[0.9]])
    

    epochs = 1000
    error_threshold = 0.002

    for i in range(epochs):
      
        z1 = np.dot(X, w1) + b1.T
        a1 = sigmoid_activation(z1)
        z2 = np.dot(a1, w2) + b2.T
        y_pred = sigmoid_activation(z2)
     
        error = y - y_pred
        delta2 = error * sigmoid_derivative(z2)
        delta1 = np.dot(delta2, w2.T) * sigmoid_derivative(z1)
        w2 += alpha * np.dot(a1.T, delta2)
        b2 += alpha * np.sum(delta2, axis=0, keepdims=True).T
        w1 += alpha * np.dot(X.T, delta1)
        b1 += alpha * np.sum(delta1, axis=0, keepdims=True).T
      
        mse = np.mean(error ** 2)
        
    
        if mse <= error_threshold:
            break
    
    return w1, b1, w2, b2, i+1


w1, b1, w2, b2, epochs = neural_network(X, y, 0.05)


print('Trained Weights and Biases of Hidden Layer:', w1, b1)
print('Trained Weights and Biases of Output Layer:', w2, b2)
print('Number of Epochs:', epochs)
```
